```python
# a) Abra o arquivo World Happiness.

import pandas as pd

caminho = './Arquivos/world_happiness_report_2015.csv'

dados = pd.read_csv(caminho)

dados
```

```python
mediaFelicidade = dados['happiness score'].mean()

# Tratando a resposta para ter 2 casas decimais
print(f'A média é {mediaFelicidade: .2f}') 
```

```python
# MUDANDO A COLUNA economy (GDP per Capita) PARA "PIB"
dados = dados.rename(columns= {'economy (GDP per Capita)':'PIB'})
```

```python
# c) Qual a soma do PIB?
dados.PIB.sum()
```

```python
# d) Qual a soma do freedom e corrupção?

dados = dados.rename(columns = {'trust (Government Corruption)':'corrupcao'})

dados[['freedom','corrupcao']].sum()
```

```python
# e) Há dados duplicados? Quantos? Verifique quais são eles?

# Aqui faremos a soma de todos os dados duplicados para assim ele mostrar a quantidade exata de dados duplicados dentro da tabela que neste caso será 6
dados.duplicated().sum()
```

```python
# Aqui será mostrado os 6 dados duplicados conforme mostrado acima
dados[dados.duplicated() == True]
```
