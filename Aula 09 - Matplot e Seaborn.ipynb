### Primeiros passos com Matplotlib

```python
# Importando as bibliotecas

import matplotlib.pyplot as plt
import numpy as np

# Criando os arrays (matrizes) para utilizar nos gráficos

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um gráfico com os valores anteriores
plt.plot(x, y)
```

## Modificando o marcador

```python
plt.plot(x,y, marker = 'X', c = 'red' )
```

### Modificando a forma como os pontos são ligados e o tipo de linha()

```python
plt.plot(
    x,
    y,
    marker = '*',
    c = 'blue', # Tipo de cor da linha
    ds = 'steps', # Forma de ligação
    ls = '--' # Estilo da linha
)
```

### Alterando a cor dos marcadores

```python
# plot = estilo do gráfico
plt.plot(
    x, # Vetor responsável pelos valores do eixo X
    y, # Vetor responsável pelos valores do eixo Y
    marker = '*', # Marcador do gráfico
    c = 'red', # Tipo de cor da linha
    ds = 'steps', # Forma de ligação
    ls = '--', # Estilo da linha
    mec = 'blue' # Cor dos marcadores
)
```

### Adicionando outros elementos ao gráfico

```python
plt.plot(x,y) # Criando o gráfico

# Adicionando o título do eixo X
plt.xlabel('Valores de X', fontsize = 12)

# Adicionando o Título do eixe Y
plt.ylabel('Valores de Y', fontsize = 12)

# Adicionando o título do gráfico 
plt.title('Título do Gráfico', fontsize = 20, c = 'green')

# Alterando as cores dos rótulos de X e Y
plt.xticks(fontsize = 8, c = 'blue')
plt.yticks(fontsize = 8, c = 'purple')
```

## Múltiplas curvas no mesmo mecanismo gráfico

```python
# Gerar dados
x = np.linspace(0, 2 * np.pi, 100) # Gerar 100 números entre 0 e 2*PI
y = np.sin(x)
z = np.cos(x)

# Curvas dos dados
plt.plot(x,y, label = 'FUNÇÃO SENO', c = '#7B0DDB')
plt.plot(x,z, label = 'FUNÇÃO COSSENO', c = '#DB1400')

# Adicionar legenda ao gráfico conforme o label que colocamos acima
plt.legend()

# ADICIONAR UM TÍTULO AO GRÁFICO
plt.title('FUNÇÕES TRIGONOMÉTRICAS')

# ADICIONAR UM TÍTULO AOS EIXOS
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')

# SALVAR O GRÁFICO COMO IMAGEM
plt.savefig('./Gráficos/funcoes-trigonometricas.png')
```

### Exemplo 1 - Gráfico de Pizza

```python
# IMPORTE DAS BIBLIOTECAS
import matplotlib.pyplot as plt

# DADOS PARA PLOTAR NO GRÁFICO
languages = ['Java','Pyhton','PHP','JavaScript','C#','C++']
popularidade = [22.2,17.6,8.8,8,7.7,6.7]


#Explosão das fatias
explosao= [0.1, 0, 0, 0.1, 0, 0]

# DEFININDO CORES
cores = ['#3F0DDB','#DB7924','#DBE90D','#DB1F0A','#26EB22','#DC26D3']

# GERAR O GRÁFICO
# labels = rótulos de categoria | autopct = rótulos numéricos
# %1.1f%% > Formatando os valores dos rótulos com 1 casa após a vírgula
# shadow =  Sombra |
plt.pie(popularidade, labels = languages, autopct= '%1.1f%%', shadow= True,
        explode = explosao, colors = cores) #'%1.1f%%' -  são a quantidade de casas decimais na hora de mostrar no gráfico
```

## Exercicio - visualização de dados matplotlib

```python
x = np.array([-3,-2.5,-2,-1,0,1,2,2.5,3])
y = np.array([-27,-15.62,-8,-1,0,1,8,15.62,27])

plt.plot(x, y,
        label = 'Função Cúbica', #Titulo da legenda
        marker = 'o',# Tipo de marcador
        color = 'magenta', #COR DA LINHA
        ls = '--', #Estilo da linha
        lw = 3) #Espessura da linha
plt.xlabel('Valores de X', fontsize = 15.5)
plt.ylabel('Valores de Y', fontsize = 15.5)
plt.title('Título do Gráfico')
plt.legend(fontsize = 20, loc = 'lower right')
```

### SEABORN

Seaborn é uma biblioteca baseada em Matplolib que traz uma interface mais atraente para os gráficos, além de facilitar a modificação de parâmetros default do
Matplotlib e permite a utilização de dataframes.

### Gráfico de dispersão

```python
# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import randn
import pandas as pd

# Geração de 200 dados aleatórios
x = randn(200)
y = randn(200)

# Criar o gráfico de dispersão
sns.scatterplot(x= x,y= y)
```

### Exemplo 2

```python
# Importando as bibliotecas
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Mapear o caminho do arquivo
caminhoArquivo = './Arquivos/sea-level.csv'

# Alocar os dados em um Dataframe
dados = pd.read_csv(caminhoArquivo)

# Fatiar o datafrmae (selecionar a partir da linha 5 (5:) e todas as colunas(:))
dados = dados.loc[6:,:]

dados.columns = [
                'Year',
                'CSIRO - Adjusted sea level (inches)',
                'CSIRO - Lower error bound (inches)',
                'CSIRO - Upper error bound (inches)',
                'NOAA - Adjusted sea level (inches)'

                ]

# Converter os dados em númericos
dados['Year'] = dados['Year'].astype(int)
dados['CSIRO - Adjusted sea level (inches)'] = dados['CSIRO - Adjusted sea level (inches)'].astype(float)
dados['CSIRO - Lower error bound (inches)'] = dados['CSIRO - Lower error bound (inches)'].astype(float)
dados['CSIRO - Upper error bound (inches)'] = dados['CSIRO - Upper error bound (inches)'].astype(float)
dados['NOAA - Adjusted sea level (inches)'] = dados['NOAA - Adjusted sea level (inches)'].astype(float)



anos = dados['Year'].to_list()
nivel = dados['CSIRO - Adjusted sea level (inches)'].to_list()

sns.scatterplot(x = anos, y = nivel)

plt.savefig('./Gráficos/Grafico-dispersao.png')
```
