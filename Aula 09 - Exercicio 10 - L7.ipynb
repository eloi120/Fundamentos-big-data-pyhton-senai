```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import os


```

```python
# Importando o arquivo excel
caminhoArquivo = './Arquivos/DadosVendaCarros.xlsx'

dados = pd.read_excel(caminhoArquivo)

dados.head()
```

```python
dados = dados.rename(columns={
    'DadosNotaFiscal':'DataNotaFiscal',
    'Coluna 2':'Fabricante',
    'Coluna 3':'Estado',
    'PrecoVenda':'ValorVenda',
    'PrecoCusto':'ValorCusto',
    'Coluna 6':'TotalDesconto',
    'Coluna 7':'CustoEntrega',
    'Coluna 8':'CustoMaoDeObra',
    'Coluna 9':'NomeCliente',
    'Coluna 10':'Modelo',
    'Coluna 11':'Cor',
    'Coluna 12':'Ano',
    'Coluna 13':'Mês'

})

dados.head()
```

```python
dados.info()
```

```python
# Removendo todos os dados nulos com o dropna
dados = dados.dropna()

dados
```

```python
dados.info()
```

```python
dados_ano = dados.groupby('Ano')['ValorVenda'].sum()

anos = dados_ano.index  # Valores de X
vendas = dados_ano.values #Valores de Y


# Criar o gráfico
plt.bar(anos,vendas)
```

### Dados de 2016

```python
# Filtrando as vendas de 2016
dados_2016 = dados[dados['Ano'] == 2016]

# Agrupar as vendas por mês
dados_2016.groupby('Mês')['ValorVenda'].sum()

```
