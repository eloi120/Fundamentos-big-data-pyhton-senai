```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
caminhoArquivo = './Arquivos/company_sales_data.csv'

dados = pd.read_csv(caminhoArquivo)

dados.head()
```

```python
listaLucro = dados['total_units'].to_list()
listaMes = dados['month_number'].to_list()
```

```python
# Construção do gráfico
plt.plot(
        listaMes, # Eixo X
        listaLucro, # Eixo Y
        label = 'Total vendido no mês', # Conteúdo da legenda 
        marker = 'o', # Cor da linha
        c = 'red' , # Marcador
        markerfacecolor = 'k', # Cor interna do marcador
        ls = '--', # Tipo de linha
        linewidth = 3 # Espessura da linha
         
    
)

plt.xlabel('Meses') # Título do eixo X
plt.ylabel('Vendas em dólares') # Título do eixo Y
plt.title('Vendas por mês do ano anterior')
plt.yticks([10000,20000,30000,40000,50000]) # Ajustando os intervalos do eixo Y
plt.xticks(listaMes)# Ajustando os rótulos do eixo X

plt.legend(loc = 'lower right')


 
dados.info()
```

## Exercício 03

```python
# Criar listas com base nas colunas do Dataframe

listaMeses = dados['month_number'].to_list()
faceCreamSalesData = dados['facecream'].to_list()
faceWashSalesData = dados['facewash'].to_list()
toothPasteSalesData = dados['toothpaste'].to_list()
bathingSoapSalesData = dados['bathingsoap'].to_list()
shampooSalesData = dados['shampoo'].to_list()
moisturizerSalesData = dados['moisturizer'].to_list()

#Gerar um gráfico de linhas para plotar cada varável descrita acima
# Tipo de marcador = o  | linewidth = 3

plt.plot(listaMeses,faceCreamSalesData, label = 'Face Cream', marker = 'o', linewidth = 3)
plt.plot(listaMeses,faceWashSalesData, label = 'facewash', marker = 'o', linewidth = 3)
plt.plot(listaMeses,toothPasteSalesData, label = 'toothpaste', marker = 'o', linewidth = 3)
plt.plot(listaMeses,bathingSoapSalesData, label = 'bathingsoap', marker = 'o', linewidth = 3)
plt.plot(listaMeses,shampooSalesData, label = 'shampoo', marker = 'o', linewidth = 3)
plt.plot(listaMeses,moisturizerSalesData, label = 'moisturizer', marker = 'o', linewidth = 3)

# Adiocionando Título aos eixos
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')

# Adicionando legenda ao gráfico
plt.legend(loc = 'upper left') # o legend pega os nomes que foram definidos anteriormente no label Ex: plt.plot(listaMeses,faceCreamSalesData, label = 'Face Cream', marker = 'o', linewidth = 3)

# Ajustando os rótulos dos eixos
plt.xticks(listaMeses)
plt.yticks([1000,2000,4000,6000,8000,10000,12000,14000,16000,18000])

# Título do gráfico
plt.title('Dados de venda')

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Mapear o caminho do arquivo

caminhoArquivo = './Arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
dados = pd.read_csv(caminhoArquivo)

#Gerar Listas
listaMeses = dados['month_number'].to_list()
vendasCremeDental = dados['toothpaste'].to_list()

# Criar o gráfico
sns.scatterplot(x = listaMeses, y = vendasCremeDental, label = 'Vendas de creme dental')
# Adicionando Grade ao Gráfico
plt.grid(True, linewidth = 0.5, linestyle = '--', c= '#0ADB9A')
# Adicionando título ao gráfico
plt.xlabel('Meses')
plt.ylabel('Unidades vendidas')
plt.title('Vendas de creme dental')
```
