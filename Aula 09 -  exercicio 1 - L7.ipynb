```python
# importar as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import os

# Mapear o caminho do arquivo
caminhoArquivo = './Arquivos/company_sales_data.csv'

# Alocar os dados em um dataframe
dados = pd.read_csv(caminhoArquivo)

# Visualizar o cabe√ßalho
dados.head()
```

```python
# Selecionar os dados que v√£o ser apresentados no gr√°fico

listaLucro = dados['total_profit'].to_list()
listaMeses = dados['month_number'].to_list()

# Criar o gr√°fico utilizando os dados listados anteriormente
plt.plot(listaMeses, listaLucro, label = 'Lucro Mensal do ano anterior')

plt.xlabel('N√∫mero do m√™s') # T√≠tulo do eixo X
plt.ylabel('Lucro') # T√≠tulo do eixo Y
plt.title('Lucro mensal da empresa') # T√≠tulo do gr√°fico
plt.legend() # Legenda do gr√°fico
```
